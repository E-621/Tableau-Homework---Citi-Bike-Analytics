```python
import pandas as pd
import os
```

```python
path = './data/'
files = os.listdir(path)
```

```python
df_list = []
for f in files:
    temp_path = path+f
    temp_df = pd.read_csv(temp_path)
    df_list.append(temp_df)
final_df = pd.concat(df_list)
final_df['gender'] = final_df['gender'].astype(str)
#final_df['gender'].replace({'1':'Male', '2':'Female','0':'Unknown'})
final_df['gender'] = final_df['gender'].replace(['1','2','0'],['Male','Female','Unknown'])
final_df
```

```python
final_df.to_csv('city_bike.csv', index = False)
```
